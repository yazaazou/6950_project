```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('en_climate_daily_AB_3075858_2001_P1D.csv')
```

```python
df.describe()
```

```python
df.columns
```

```python
df=df.rename(columns={"Mean Temp (°C)": "meanTemp"})
```

```python
plt.figure(figsize=(12,8))
plt.plot(df.meanTemp,'.')
plt.xlabel('Day of year')
plt.ylabel('Mean Temp. Celcius')
plt.title('Mean Temp Dist for Sexsmith year 2001')
plt.savefig('Sexsmith.jpg')
plt.show()
```
